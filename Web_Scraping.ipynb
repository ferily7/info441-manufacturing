```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs, SoupStrainer as ss
import requests as r
```

```python
websites = ['https://www.newegg.com/CPUs-Processors/Category/ID-34',
            'https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38',
            'https://www.newegg.com/Motherboards/Category/ID-20']
product_types = ['CPU Processors', 'Graphics Cards', 'Motherboards']

item_title = []
price = []
product_type = []
brand = []

for index in range(len(websites)): 
    
    page = r.get(websites[index], verify = False)
    content = bs(page.content, 'html.parser')
    
    # Get <div> tags for item info
    item_info = content.findAll('div', {'class': 'item-info'})

    for info in item_info:
        title_tag = info.findAll('a', {'class': 'item-title'})
        for tag in title_tag:
            title = tag.contents[1]
            item_title.append(title)
            brand_name = title.split(' ', 1)[0]
            brand.append(brand_name)

        price_tag = info.findAll('li', {'class': 'price-current'})
        for tag in price_tag:
            price_value = tag.find('strong').contents[0]
            price.append(price_value + '.99')

        product_type.append(product_types[index])

    # print(item_title)
    # print(price)

# Create dictionary 
dictionary = [('Title', item_title), ('Price', price), ('Product Type', product_type), ('Brand', brand)]

# Create dataframe using dictionary
df = pd.DataFrame.from_items(dictionary)
    
df
```

```python
# Write the dataframe to a .csv file in this directory (make sure to exclude row numbers)
df.to_csv('products.csv', index = False)
```
